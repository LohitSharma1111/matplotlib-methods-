# pyplot method is for regular datasets where as OO method is used for advanced plottinng 

```python
heart=pd.read_csv("heart-disease.csv")
heart
```

```python
heart[:10]
```

```python

Bfifty=heart[heart["chol"]<300]
Bfifty.tail()
```

```python
#normal pyplot method (PLOTTING DIRECTLY FROM DATAFRAMES)
Bsixty.plot(kind="scatter",
           x="age",
           y="chol",
           c="target",figsize=(5,5));
```

```python
#OO method

import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(5,5))
heart.plot(kind="scatter",
          x="age",
          y="chol",
          c="target",
          ax=ax);
ax.set_xlim([45,10]);
```

```python
Afifty=heart[heart["age"]>70]
Afifty.tail()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
fig,ax=plt.subplots(figsize=(5,5))

#plotting the data 
scatter=ax.scatter(
                  x=Bfifty["chol"],
                  y=Bfifty["age"],
                  c=Bfifty["target"])

#customizing the plot
ax.set(title="cholestrol below 150",
              xlabel="age",
              ylabel="chol");

#adding TARGET slot in the figure based on the column name in the dataframe
ax.legend(*scatter.legend_elements(),title="target");

#adding a horizontal line in  a fig based on their values 

ax.axhline(Bfifty["age"].mean(),
          linestyle="--");
```
